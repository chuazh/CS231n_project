```python
import matplotlib.pyplot as plt
import numpy as np

y = np.loadtxt('KT_model_data/vis_gt.dat',delimiter = ',')
scores = np.loadtxt('KT_model_data/vis_scores.dat', delimiter = ',')

plt.figure()
for i in range(14):
    plt.subplot(14,1,i+1)
    plt.plot(scores[:,i],'r')
    plt.plot(y[:,i],'b')

import pickle

f = open('KT_model_data/train_loss.dat','rb')
train_loss = pickle.load(f)
f.close()
f = open('KT_model_data/val_loss.dat','rb')
val_loss = pickle.load(f)
f.close()

plt.figure()
plt.plot(train_loss)
plt.plot(val_loss)
```
